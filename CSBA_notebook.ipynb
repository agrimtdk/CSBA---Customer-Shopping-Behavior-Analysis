```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.info()
```

```python
#creating a distinguishing table for age groups
age_label = ['Young Adult', 'Adult', 'Middle Aged', 'Senior Citizen']
df['age_group'] = pd.qcut(df['age'], q= 4, labels = age_label)
```

```python
df.head(5)
```

```python
#creating col for freq of purchasing days
freq_mapping = {
    'Fortnightly': 14, 'Weekly': 7, 'Monthly': 30, 'Quarterly': 90, 'Biweekly': 14, 'Annually': 365, 'Every 3 months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(freq_mapping)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis= 1)
```
